### Clustering using sklearn library

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(data)
```

```python
centers = kmeans.cluster_centers_
centersx = centers[:,0]
centersy =  centers[:, 1]
plt.scatter(datax, datay)
plt.scatter(centersx,centersy, marker='o', lw=5)
```

#### Initializing Centers

```python
# random uniformly generated points from 
### Centers generating methods
# centers = np.random.uniform([datax.min(), datay.min()], [datax.max(), datay.max()], size=(3,2))
# centers = np.random.uniform(data.min(axis=0), data.max(axis=0), size=(3,2))

mykmeans = KMeansMy(n_centers=3)
# centers = KMeansMy(n_centers=3).generate_centers_uniformly(data)
centers = mykmeans.generate_centers_sampling(data)
### END Centers generating methods

# Plotting the centers and data points
plt.scatter(data[:,0], data[:, 1])
plt.scatter(centers[:,0], centers[:, 1], marker='o', lw=5)
```

#### Calculating distance

```python
def norm(centers, data): # norm_c0 = sqrt[(x0-c0)^2 + (x1-c0)^2 + (x2-c0)^2 ... ] & similar for c1, c2, ...  
    dist = []
    for center in centers:
        diff = data - center # difference between center and data point
        sqrd = np.square(diff) # elementwise squaring 
        sumd = np.sum(sqrd, axis=1) # not summing all data. but only n_features,i.e. columns (axis = 1)
        sqrt = np.sqrt(sumd) # square root after sum
        dist.append(sqrt) # collecting all the sum
#         print(sqrt, sqrt.shape)
        
    return np.array(dist)

distances = norm(centers, data)

color_list = ['r', 'g', 'y']
    
plt.scatter(data[:,0], data[:, 1])
plt.scatter(centers[:,0], centers[:, 1], marker='o', c=color_list,lw=5)
```

#### Finding smallest distance center

```python
print(distances.shape)
belongto = np.argmin(distances, axis = 0) # finding minimum over the three centers
print(belongto)
```

```python
# these points belong to corresponding centers now
# we should assign these point to corresponding centers

plt.scatter(centers[:,0], centers[:, 1], marker='o', c=color_list,lw=5)
plt.scatter(datax, datay, marker = '.', c=belongto, cmap=matplotlib.colors.ListedColormap(color_list))
```

#### Update the centers for the cluster

```python
belongto == 0 # binary mask of the logic elementwise
```

```python
for index in range(len(centers)):
    centers[index] = np.mean(data[belongto == index], axis=0) # mean among the row .: axis=0
centers
```

```python
plt.scatter(centers[:,0], centers[:, 1], marker='o', c=color_list,lw=5)
plt.scatter(datax, datay, marker = '.', c=belongto, cmap=matplotlib.colors.ListedColormap(color_list))
```
